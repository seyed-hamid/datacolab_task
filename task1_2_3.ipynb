```python
import pandas as pd
import json
import re

df = pd.read_csv("/content/to_fill.csv")

df["full_text"] = ""
df['start'] = ""
df['end'] = ""
df['last_words'][1] = 'all the way down to the ocean.'
df['last_words'][12] = 'name or what led to that fall.'
for index, row in df.iterrows():
    json_path = row["source_video_id"]

    with open(str(json_path) + ".json") as f:
        json_data = json.load(f)
    
    text = json_data["text"]
    word = json_data['words']
    start_words = row["first_words"].split()
    end_words = row["last_words"].split()

    bit_index = None
    for i, item in enumerate(word):
      if item["text"] == start_words[0]:
        bit_index = i
        if bit_index is not None and bit_index < len(word) - 2:
            if word[bit_index + 1]["text"] == start_words[1] and word[bit_index + 2]["text"] == start_words[2]:
               df.at[index, "start"] = word[bit_index]["start"]
               bit_end_time = word[-1]["end"]
               #print(f"The word '{start_words[0]}' starts at time {word[bit_index]['start']} and ends at time {bit_end_time}.")
               break
              
        else:
          print(f"The word '{start_words[0]}' does not appear in the data list or does not have at least two words following it.")
    bit_index_end = None  
    for j, items in enumerate(word):
      if items["text"] == end_words[-1]:
        bit_index_end = j
        if bit_index_end is not None and bit_index_end < len(word):
            if word[bit_index_end - 1]["text"] == end_words[-2] and word[bit_index_end - 2]["text"] == end_words[-3]:
               #df.at[index, "start"] = word[bit_index]["start"]
               #bit_end_time = word[-1]["end"]
               df.at[index, "end"] =  word[bit_index_end]["end"]
               #print(f"The word '{end_words[-1]}' starts at time {word[bit_index_end]['start']} and ends at time {bit_end_time}.")
               break
              
        else:
          print(f"The word '{end_words[-1]}' does not appear in the data list or does not have at least two words following it.")
      


    start_index = re.search(re.escape(row["first_words"]), text).start()
    end_match = re.search(re.escape(row["last_words"]), text)
    if end_match:
        end_index = end_match.end()
    else:
        end_index = None
        print(row["last_words"])

    print("Start index:", start_index)
    print('words = ' ,text[start_index] )
    print("End index:", end_index)
    print("Start index:", start_index)
    print("End index:", end_index)
    extracted_text = text[start_index:end_index]
    df.at[index, "full_text"] = extracted_text
```
